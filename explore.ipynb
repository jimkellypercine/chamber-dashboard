```python
import matplotlib.pylab as plt
import pandas as pd
import numpy as np
```

```python
df_event_attendance = pd.read_csv('/Users/jimkellypercine/Desktop/Chamber_Project/data/Event_Attendance_Members.csv')
df_event_attendance.head()
df_event_attendance.columns = [c.replace('-', '_')for c in df_event_attendance.columns]
df_event_attendance.columns = [c.replace(' ', '_')for c in df_event_attendance.columns]
df_event_attendance.head()
```

```python
member_Names = df_event_attendance["Member_Name"]

member_Names.head(10)

top_spenders = df_event_attendance.nlargest(10, 'Fees')[['Member_Name', 'Fees']]
top_spenders.plot(kind='bar', x='Member_Name', y='Fees', legend=False)
plt.ylabel('Fees')
plt.title('Top 10 Members by Fees')
plt.xticks(rotation=45, ha='right')
plt.show()

#this chart shows the top 10 members by fees paid to the chamber

total_fees = df_event_attendance['Fees'].sum()
average_attendance = df_event_attendance['Attended'].mean()
print('Total Fees: ', total_fees)
print('Average Attendance: ', average_attendance)

top_spenders.set_index('Member_Name').plot(kind='pie', y='Fees', autopct='%1.1f%%', legend=False)
plt.ylabel('')
plt.title('Top 10 Members by Fees (Pie Chart)')
plt.show()
```
